```python
import numpy as np
```

```python
normal_data=np.random.normal(loc=50,scale=3,size=100)
normal_data
```

```python
o=np.array([150,200,250])
o_data=np.concatenate([normal_data,o])
print(o)
print(o_data)
```

```python
nm=np.mean(normal_data)
on=np.mean(o_data)
print(nm)
print(on)
```

```python
import pandas as pd
```

```python
s=pd.DataFrame({"data_normal":normal_data})
s
```

```python
s1=pd.DataFrame({"data_outlier":o_data})
s1
```

```python
s.kurt()
```

```python
s1.kurt()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)

sns.histplot(normal_data,kde=True)
plt.title("normal_data")
plt.subplot(1,2,2)

sns.histplot(o_data,kde=True)
plt.title("o_data")
plt.show()

from scipy.stats import skew
skew(normal_data)
skew(o_data)

```
