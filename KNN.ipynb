```python
#feature engineering
x=d[["danceability","energy","tempo","loudness","Valence"]]
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss
```

```python
x=ss.fit_transform(x)
x
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(n_neighbors=6,metric="euclidean")
model.fit(x)
```

```python
#d["singer"].unique()
```

```python
x=["danceability","energy","tempo","loudness","Valence"]
def rec(song):
    x_index=d[d["song_name"]==song].index[0]
    query_point=d.loc[[x_index],x].values
    distance,indices=model.kneighbors(query_point)
    return d.iloc[indices[0]]["song_name"]
#3.test it
print(rec("Andhi Mazhai Pozhigirathu"))
```
