```python
import pandas as pd
```

```python
d=pd.read_csv("StudentsPerformance.csv")
d
```

```python
d.columns
```

```python
import matplotlib.pyplot as plt
```

```python
d.groupby("gender")[["math score", "reading score",
       "writing score"]].mean().plot(kind="bar")
plt.title("genter vs score")
plt.xlabel("gender")
plt.ylabel("score")
plt.legend()
plt.show()
```

```python
c=["5 star","dairy milk","kitkat"]
s=[200,400,350]
plt.bar(c,s,color=["yellow","red","black"])
plt.xlabel("chocolate company")
plt.ylabel("sales in quandity")
plt.show()
```

```python
plt.hist(d["math score"],bins=15)
plt.title("distribution of math score")
plt.xlabel("score")
plt.ylabel("freqency")
plt.show()
```

```python
plt.scatter(d["math score"],d["reading score"])
plt.title("math score vs reading score")
plt.xlabel("math score")
plt.ylabel("reading score")
plt.show()
```

```python
 #line plot
avg_s=d[["math score",'reading score',"writing score"]].mean()
avg_s
```

```python
x=avg_s.index
y=avg_s.values
```

```python
plt.plot(x,y)
plt.title("maths vs reading")
plt.xlabel("math score")
plt.ylabel("reading score")
plt.show()
```
