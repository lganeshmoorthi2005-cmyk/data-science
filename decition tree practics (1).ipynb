```python
import pandas as pd

# Data
data = {
    'popcorn': [1,1,0,0,1,1,0],
    'soda':    [1,0,1,1,1,0,0],
    'age':     [7,12,18,35,38,50,83],
    'movie':   [0,0,1,1,1,0,0]
}

df = pd.DataFrame(data)
print(df)
```

```python
x = df[['popcorn', 'soda', 'age']]  # Features
y = df['movie']                     # Target
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion='gini') 
model.fit(x, y)
```

```python
# Example prediction
new_data = [[1, 1, 25]]  # popcorn=1, soda=1, age=25
prediction = model.predict(new_data)

print("Prediction:", prediction)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plot_tree(model, feature_names=['popcorn','soda','age'], class_names=['No','Yes'], filled=True)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l=LabelEncoder()
c["video_id"]=l.fit_transform(c['video_id'])
c["video_id"]
```

```python
c
```

```python
x=c[["user_id","video_id","video_duration"]]
y=c["watch_time"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(criterion="gini",max_depth=3)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 3))
plot_tree(
    model,
    feature_names=c.columns.tolist(),          # ✅ column names from the DataFrame
    class_names=[str(c) for c in model.classes_],  # ✅ class labels from the model
    filled=True
)
plt.show()
```
